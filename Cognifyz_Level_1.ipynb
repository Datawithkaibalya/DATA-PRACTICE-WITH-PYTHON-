This notebook covers data exploration, preprocessing, and descriptive analysis.

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
df = pd.read_csv("Dataset .csv")
df.head()
```

```
# Dataset shape and info
df.shape, df.info()
```

```
# Missing values
df.isnull().sum()
```

```
# Handling missing values
num_cols = df.select_dtypes(include=np.number).columns
cat_cols = df.select_dtypes(exclude=np.number).columns

df[num_cols] = df[num_cols].fillna(df[num_cols].median())
df[cat_cols] = df[cat_cols].fillna(df[cat_cols].mode().iloc[0])
```

```
# Target variable distribution
plt.hist(df["Aggregate rating"], bins=20)
plt.xlabel("Rating")
plt.ylabel("Count")
plt.title("Distribution of Aggregate Rating")
plt.show()
```

```
# Descriptive statistics
df.describe()
```

```
# Top cities and cuisines
df["City"].value_counts().head(10)
df["Cuisines"].value_counts().head(10)
```

```
# Table booking & online delivery analysis
df["Has Table booking"].value_counts(normalize=True) * 100
df.groupby("Has Table booking")["Aggregate rating"].mean()
```
