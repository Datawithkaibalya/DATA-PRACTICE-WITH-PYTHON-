```python
import pandas as pd

employee = {
    "emp_id": [101,102,103,104,105,106,107,108,109,110],
    "name": ["Amit","Priya","Rahul","Ananya","Suresh","Neha","Karan","Pooja","Vikram","Riya"],
    "age": [28,32,25,29,35,27,31,26,38,24],
    "gender": ["M","F","M","F","M","F","M","F","M","F"],
    "department": ["Data Science","HR","Data Analyst","Data Science","Manager",
                   "Data Analyst","ML Engineer","HR","Manager","Intern"],
    "salary": [65000,48000,52000,72000,90000,50000,80000,45000,95000,20000],
    "experience": [3,7,2,4,10,2,6,1,12,0],
    "city": ["Bangalore","Hyderabad","Pune","Bangalore","Chennai",
             "Hyderabad","Bangalore","Pune","Delhi","Bangalore"],
    "join_date": ["2021-06-15","2018-03-20","2022-01-10","2020-09-05",
                  "2015-11-12","2022-07-01","2019-02-18","2023-04-25",
                  "2012-08-30","2024-01-10"]}

df = pd.DataFrame(employee)
df
```

```python
def salary_category(salary):
    if salary < 40000:
        return 'Low'
    elif salary <= 70000:
        return 'Medium'
    else:
        return 'High'

df['salary_category'] = df['salary'].apply(salary_category)
```
